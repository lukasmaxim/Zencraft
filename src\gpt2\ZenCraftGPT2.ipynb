## Request Files for Input and Corpus Building 

```
    import requests
    corpusUrl = 'https://raw.githubusercontent.com/ssaltzen/Testing-Collab/master/corpus.txt'
    inputUrl = 'https://raw.githubusercontent.com/ssaltzen/Testing-Collab/master/input.txt'
    cPage = requests.get(corpusUrl)
    iPage = requests.get(inputUrl)
    corpus = cPage.text
    input = iPage.text
```

# Set up GPT2

```
tempCorpus = "tempCorpus.txt"
with open(tempCorpus, 'w') as f:
		f.write(corpus)

gpt2.download_gpt2(model_name="124M")

sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset= tempCorpus,
              model_name='124M',
              steps=1000,
              )
```
